```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt

np.random.seed(42)

for dim in (2,3,10,100,1000,10000):
  points = np.random.rand(1000, dim)  # draw 1000 vectors of size dim

  # compute the L2 distance between each pair of vectors (we want the output
  # to be a vector because that's what the next plot function expects)
  dist = euclidean_distances(points, points).reshape(points.shape[0] ** 2)

  # can also use numpy
  # dist = np.hstack([np.sqrt(np.sum((points-p)**2,axis=1)) for p in points])

  # we plot, for each possible distance value, the number of pair of points which are at that distance
  plt.hist(dist, bins="auto")
  plt.title("Dim=%d" % (dim))
  plt.show()
```
